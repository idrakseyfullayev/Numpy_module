


### üìå Kod:

```python
a = np.array([[1, 2],
              [3, 4]])

def getRange(x, y):
    return 100 * x + y

a = np.fromfunction(getRange, (2, 2))
```

---

### üîç Addƒ±m 1: `getRange(x, y)` funksiyasƒ±

Bu funksiya **iki indeks** (`x`, `y`) alƒ±r v…ô a≈üaƒüƒ±dakƒ± kimi hesablayƒ±r:

```python
return 100 * x + y
```

Y…ôni:

* `x` ‚Äì satƒ±r indeksi,
* `y` ‚Äì s√ºtun indeksi,
  v…ô n…ôtic…ô **100 \* satƒ±r + s√ºtun** olur.

---

### üîç Addƒ±m 2: `np.fromfunction(getRange, (2, 2))`

Bu kod NumPy-ya deyir:
‚Äú**(2, 2)** √∂l√ß√ºl√º bir massiv yarat v…ô h…ôr bir `(i, j)` indeksl…ôri √º√ß√ºn `getRange(i, j)` funksiyasƒ±nƒ± i≈ü…ô sal.‚Äù

---

### üí° ∆èsas m…ôqam:

`np.fromfunction()` avtomatik olaraq `x` v…ô `y` √º√ß√ºn a≈üaƒüƒ±dakƒ± kimi indeks massivl…ôri yaradƒ±r:

```python
x = [[0. 0.]
     [1. 1.]]
     
y = [[0. 1.]
     [0. 1.]]
```

---

### ‚úÖ ƒ∞ndi `getRange(x, y)` i≈ül…ôyir:

```python
100 * x + y =
[[100*0 + 0, 100*0 + 1], 
 [100*1 + 0, 100*1 + 1]] =
[[  0,   1],
 [100, 101]]
```

---

### üîö N…ôtic…ô:

```python
a = np.array([[  0.,   1.],
              [100., 101.]])
```

> **Qeyd:** N…ôtic…ôd…ô `float` √ßƒ±xƒ±r, √ß√ºnki `fromfunction()` default olaraq `dtype=float` istifad…ô edir (d…ôyi≈üm…ôk √º√ß√ºn `dtype=int` …ôlav…ô ed…ô bil…ôrs…ôn).
