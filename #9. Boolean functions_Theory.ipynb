## Boolean operations and functions

---

* **np.greater()** – İki massiv elementlərini müqayisə edərək, birinci massivdəki elementin ikincidən böyük olub-olmadığını `True/False` kimi qaytarır.
  `result = np.greater([3, 5, 2], [2, 5, 4])`

---

* **np.less()** – İki massiv elementlərini müqayisə edərək, birinci massivdəki elementin ikincidən kiçik olub-olmadığını `True/False` kimi qaytarır.
  `result = np.less([3, 5, 2], [2, 5, 4])`

---

* **np.equal()** – İki massivdəki uyğun elementlərin bərabər olub-olmadığını `True/False` kimi qaytarır.
  `result = np.equal([1, 2, 3], [1, 4, 3])`

---

* **np.array\_equal()** – İki massiv tamamilə eyni forma və eyni elementlərə malikdirsə, `True` qaytarır, əks halda `False`.
  `result = np.array_equal(np.array([1, 2]), np.array([1, 2]))`

---

* **np.any()** – Verilmiş massivdə heç olmasa bir `True` (və ya sıfırdan fərqli) element varsa, `True` qaytarır.
  `result = np.any([0, 0, 3])`

---

* **np.any(a > 5)** – Massivdə 5-dən böyük heç olmasa bir element varsa, `True` qaytarır.
  `result = np.any(a > 5)`

---

* **np.all()** – Verilmiş massivdəki bütün elementlər `True` (və ya sıfırdan fərqli) olarsa, `True` qaytarır, yoxsa `False`.
  `result = np.all([1, 2, 3])`

---

* **np.all(a > 5)** – Massivdəki bütün elementlər 5-dən böyükdürsə, `True` qaytarır, əks halda `False`.
  `result = np.all(a > 5)`

---

* **np.isinf()** – Massivin elementlərinin sonsuz (`inf` və ya `-inf`) olub-olmadığını `True/False` kimi göstərir.
  `result = np.isinf(np.array([1, np.inf, -np.inf, 5]))`

---

* **np.isnan()** – Massivin elementlərinin `NaN` (Not a Number) olub-olmadığını `True/False` kimi göstərir.
  `result = np.isnan(np.array([1, np.nan, 2]))`

---

* **np.isfinite()** – Massivin elementlərinin sonlu (yəni nə `NaN`, nə də `±inf`) olub-olmadığını `True/False` kimi göstərir.
  `result = np.isfinite(np.array([1, np.nan, np.inf, 5]))`

---

* **np.iscomplex()** – Massivin elementlərinin kompleks ədəd olub-olmadığını `True/False` kimi göstərir.
  `result = np.iscomplex(np.array([1, 2 + 3j, 4]))`

---

* **np.isreal()** – Massivin elementlərinin real ədəd olub-olmadığını `True/False` kimi göstərir.
  `result = np.isreal(np.array([1, 2 + 3j, 4]))`

---

* **np.logical\_and()** – İki massiv üzrə məntiqi AND (və) əməliyyatı aparmaq üçün istifadə olunur. Hər iki massivdə uyğun indekslərdəki elementlər `True` olduqda nəticə `True` olur.
  `result = np.logical_and([True, False, True], [False, False, True])`

---

* **np.logical\_or()** – İki massiv üzrə məntiqi OR (və ya) əməliyyatı aparmaq üçün istifadə olunur. İki massivdən hər hansı birində `True` varsa nəticə `True` olur.
  `result = np.logical_or([True, False, True], [False, False, True])`

---

* **np.logical\_not()** – Massivin hər bir elementinə məntiqi NOT (inkar) əməliyyatı tətbiq edir. `True` → `False`, `False` → `True`.
  `result = np.logical_not([True, False, True])`

---

* **np.logical\_xor()** – İki massiv üzrə məntiqi XOR (eksklüziv və ya) əməliyyatı aparmaq üçün istifadə olunur. Yalnız birində `True` olduqda nəticə `True` olur, hər ikisində eyni olduqda `False`.
  `result = np.logical_xor([True, False, True], [False, False, True])`
