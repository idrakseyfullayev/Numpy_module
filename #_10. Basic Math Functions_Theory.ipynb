## Basic Math Functions

---

* **np.sum()** – Massivin bütün elementlərinin cəmini və ya müəyyən `axis` üzrə cəmini hesablamaq üçün istifadə olunur.
  `result = np.sum(arr)`
  `result_axis = np.sum(arr, axis=0)`

---

* **np.mean()** – Massivin elementlərinin ortalamasını (average) və ya müəyyən `axis` üzrə ortalamasını hesablamaq üçün istifadə olunur.
  `result = np.mean(arr)`
  `result_axis = np.mean(arr, axis=1)`

---

* **np.max()** – Massivin ən böyük elementini və ya müəyyən `axis` üzrə maksimum dəyəri tapmaq üçün istifadə olunur.
  `result = np.max(arr)`
  `result_axis = np.max(arr, axis=0)`

---

* **np.min()** – Massivin ən kiçik elementini və ya müəyyən `axis` üzrə minimum dəyəri tapmaq üçün istifadə olunur.
  `result = np.min(arr)`
  `result_axis = np.min(arr, axis=1)`

---

* **np.abs()** – Massivin elementlərinin mütləq dəyərlərini hesablamaq üçün istifadə olunur.
  `result = np.abs(np.array([-1, -2, 3]))`

---

* **np.amax()** – Massivin ən böyük elementini və ya müəyyən `axis` üzrə maksimum dəyəri tapmaq üçün istifadə olunur (np.max() ilə eyni funksionallıq).
  `result = np.amax(arr)`
  `result_axis = np.amax(arr, axis=0)`

---

* **np.amin()** – Massivin ən kiçik elementini və ya müəyyən `axis` üzrə minimum dəyəri tapmaq üçün istifadə olunur (np.min() ilə eyni funksionallıq).
  `result = np.amin(arr)`
  `result_axis = np.amin(arr, axis=1)`

---

* **np.around()** – Massivin elementlərini müəyyən sayda onluq dəqiqliklə yuvarlaqlaşdırmaq üçün istifadə olunur.
  `result = np.around(arr, decimals=2)`

---

* **np.argmax()** – Massivin ən böyük elementinin indeksini və ya müəyyən `axis` üzrə maksimum elementin indeksini tapmaq üçün istifadə olunur.
  `index = np.argmax(arr)`
  `index_axis = np.argmax(arr, axis=0)`

---

* **np.argmin()** – Massivin ən kiçik elementinin indeksini və ya müəyyən `axis` üzrə minimum elementin indeksini tapmaq üçün istifadə olunur.
  `index = np.argmin(arr)`
  `index_axis = np.argmin(arr, axis=1)`

---

* **np.linspace()** – Başlanğıc (`start`) və son (`stop`) nöqtələr arasında `num` sayda bərabər aralıqla ədədlərdən ibarət massiv yaratmaq üçün istifadə olunur.
  `result = np.linspace(start=0, stop=1, num=5)`

---

* **np.sin()** – Massivin elementlərinə sinus funksiyasını tətbiq etmək üçün istifadə olunur.
  `result = np.sin(np.array([0, np.pi/2, np.pi]))`

---

* **np.cos()** – Massivin elementlərinə kosinus funksiyasını tətbiq etmək üçün istifadə olunur.
  `result = np.cos(np.array([0, np.pi/2, np.pi]))`

---

* **np.random.rand()** – Verilmiş ölçüdə (shape) 0, 1 intervalında bərabər ehtimallı təsadüfi ədədlərdən ibarət massiv yaratmaq üçün istifadə olunur.
  `result = np.random.rand(3, 2)`  # 3 sıra, 2 sütunlu massiv

---

* **np.random.randint()** – Verilən aralıqda tam ədədlərdən ibarət təsadüfi massiv yaratmaq üçün istifadə olunur.
  `result = np.random.randint(low=0, high=10, size=(3, 4))`  # 0-dan 9-a qədər (10 daxil deyil) 3x4 ölçüdə massiv

---

* **np.random.randn()** – Standart normal paylanmaya (orta 0, dispersiya 1) uyğun təsadüfi ədədlərdən ibarət massiv yaratmaq üçün istifadə olunur.
  `result = np.random.randn(3, 2)`  # 3 sıra, 2 sütunlu massiv

---

* **np.random.pareto()** – Pareto paylanmasından təsadüfi ədədlər yaratmaq üçün istifadə olunur.
  `result = np.random.pareto(a=3.0, size=(3, 4))`  # `a` - Pareto paylanmasının shape parametri, 3x4 ölçülü massiv

---

* **np.random.beta()** – Beta paylanmasından təsadüfi ədədlər yaratmaq üçün istifadə olunur.
  `result = np.random.beta(a=2.0, b=5.0, size=(3, 4))`  # `a` və `b` - Beta paylanmasının shape parametrləri, 3x4 ölçülü massiv

---

* **np.random.normal()** – Normal paylanmadan (Gauss paylanması) təsadüfi ədədlər yaratmaq üçün istifadə olunur.
  `result = np.random.normal(loc=0.0, scale=1.0, size=(3, 4))`  # `loc` — orta, `scale` — standart sapma, 3x4 ölçülü massiv

---

* **np.random.uniform()** – Verilmiş intervalda bərabər ehtimallı təsadüfi ədədlər yaratmaq üçün istifadə olunur.
  `result = np.random.uniform(low=0.0, high=1.0, size=(3, 4))`  # `[low, high)` aralığında, 3x4 ölçülü massiv

---


* **np.random.seed()** – Təsadüfi ədədlərin təkrarlanabilirliyini təmin etmək üçün random generatora başlanğıc dəyər (tohum) təyin etmək üçün istifadə olunur.
  `np.random.seed(42)`  # Eyni nəticələr üçün eyni seed dəyəri istifadə edilir

---

* **np.random.shuffle()** – Massivin elementlərini yerində qarışdırmaq (shuffle) üçün istifadə olunur.
  `np.random.shuffle(arr)`  # `arr` massivinin elementləri təsadüfi sıraya düzülür (yerində dəyişir). Orijinal array-in özünü qarışdırır, geri dönüş yoxdur

---

* **np.random.permutation()** – Verilmiş massivdə və ya ədədlər aralığında elementləri təsadüfi şəkildə sıralayaraq yeni massiv yaratmaq üçün istifadə olunur.  Yeni qarışdırılmış nüsxə verir, orijinal qalmır.
  `result = np.random.permutation(arr)`  # `arr` massivinin elementləri qarışdırılmış yeni massiv qaytarılır
  və ya
  `result = np.random.permutation(5)`  # 0-dan 4-ə qədər ədədlərin qarışdırılmış siyahısı

---

* **np.median()** – Massivin orta dəyərini (medianını) hesablamaq üçün istifadə olunur.
  `result = np.median(arr)`
  `axis` parametri ilə müəyyən istiqamətdə medianı hesablamaq mümkündür:
-  `np.median(arr, axis=0)` – sütunlar üzrə
-  `np.median(arr, axis=1)` – sətirlər üzrə

---

* **np.var()** – Massivin dispersiyasını (variansını) hesablamaq üçün istifadə olunur.
  `result = np.var(arr)`
  `axis` parametri ilə müəyyən istiqamətdə varians hesablana bilər:
-  `np.var(arr, axis=0)` – sütunlar üzrə
- `np.var(arr, axis=1)` – sətirlər üzrə

---

* **np.std()** – Massivin standart sapmasını hesablamaq üçün istifadə olunur.
  `result = np.std(arr)`
  `axis` parametri ilə müəyyən istiqamətdə standart sapma hesablana bilər:
-  `np.std(arr, axis=0)` – sütunlar üzrə
-  `np.std(arr, axis=1)` – sətirlər üzrə

---

* **np.corrcoef()** – İki və ya daha çox massiv arasında korrelyasiya (Pearson correlation coefficient) matrisini hesablayır.
  `result = np.corrcoef(arr1, arr2)`
-   Giriş massivləri eyni ölçüdə olmalıdır.
  `np.corrcoef(arr1, arr2)[0, 1]` – arr1 və arr2 arasında korrelyasiya dəyərini verir.
  

```
[[1,        corr(arr1, arr2)],
 [corr(arr2, arr1),    1     ]]
```

- Qısa izahı:

1. `[0, 0]` — `arr1`-in öz-özünə korrelyasiyası (həmişə 1-dir)
2. `[0, 1]` — `arr1` və `arr2` arasındakı korrelyasiya əmsalı (correlation coefficient)
3. `[1, 0]` — `arr2` və `arr1` arasındakı korrelyasiya (eyni ilə `[0,1]`)
4. `[1, 1]` — `arr2`-nin öz-özünə korrelyasiyası (həmişə 1-dir)

- **Qeyd:** Korrelyasiya -1 ilə 1 arasında dəyişir və dəyişənlər arasındakı xətti əlaqənin gücünü və istiqamətini göstərir.

- Yəni:

* 1 — mükəmməl müsbət korrelyasiya,
* 0 — korrelyasiya yoxdur,
* -1 — mükəmməl mənfi korrelyasiya.


- Sadəcə desək:
**`np.corrcoef`** dəyişənlər arasındakı əlaqənin **gücünü və istiqamətini** ölçür,
**`np.cov`** isə dəyişənlər arasındakı **birlikdə dəyişmə səviyyəsini** (kovariasiyanı) göstərir.
  

---

* **np.cov()** – Kovariasiya matrisini hesablamaq üçün istifadə olunur.
  `result = np.cov(arr1, arr2)`
-  Giriş massivləri 1D və ya 2D ola bilər.
-  Nəticə matrisin `[0, 1]` və ya `[1, 0]` indekslərində iki dəyişən arasındakı kovariasiyanı göstərir.

- **Qeyd:** Varsayılan olaraq, sətirlər müşahidələr, sütunlar dəyişənlər olmalıdır (`rowvar=True`). Əgər `arr1` və `arr2` 1D-dirsə, `np.cov(arr1, arr2)[0, 1]` onların kovariasiya dəyərini verir.



`np.cov(arr1, arr2)` funksiyasının nəticəsi 2x2 ölçülü kovariasiya matrisi olur və dörd dəyər çıxır:

```
[[var(arr1), cov(arr1, arr2)],
 [cov(arr2, arr1), var(arr2)]]
```

Qısa izahı:

1. `[0, 0]` — `arr1`-in dispersiyası (variance)
2. `[0, 1]` — `arr1` və `arr2` arasındakı kovariasiya (covariance)
3. `[1, 0]` — `arr2` və `arr1` arasındakı kovariasiya (covariance), yəni `[0, 1]` ilə eyni
4. `[1, 1]` — `arr2`-nin dispersiyası (variance)

Beləliklə, matrisdə diaqonal elementlər dəyişənlərin öz dispersiyalarını, digər elementlər isə dəyişənlər arasındakı kovariasiyanı göstərir.


---

* **np.correlate()** – İki 1D massiv arasında korrelyasiya (cross-correlation) əməliyyatını yerinə yetirir.
  `result = np.correlate(a, v, mode='valid')`
  Burada:

  * `a`, `v` – 1D massivlər
  * `mode` – çıxış ölçüsünü təyin edir (`'valid'`, `'same'`, `'full'`)


### ✅ `np.corrcoef(x, y)` → **Pearson korrelyasiya əmsalı**

**Riyazi düstur:**

$$
\rho_{xy} = \frac{\text{cov}(x, y)}{\sigma_x \cdot \sigma_y}
$$

Burada:

* $\text{cov}(x, y)$ — kovariasiya
* $\sigma_x, \sigma_y$ — standart sapmalar
* Nəticə həmişə **\[-1, 1]** arasında olur



### ✅ `np.correlate(x, y)` → **Discrete cross-correlation**

**Riyazi düstur (dot product):**

$$
(x \star y)[n] = \sum_{i} x[i] \cdot y[i+n]
$$

> Bu, vektorların sürüşdürülərək üst-üstə düşən hissələrinin **çoxlu nöqtə hasilinin** cəmini hesablayır.

Əsasən **signal processing** üçündür, statistik mənada **korrelyasiya deyil**.



### Qısa müqayisə:

| Funksiya    | Düstur                                             | Sahə         |
| ----------- | -------------------------------------------------- | ------------ |
| `corrcoef`  | $$\frac{\text{cov}(x, y)}{\sigma_x \cdot \sigma_y}$$ | Statistika   |
| `correlate` | $$\sum x[i] \cdot y[i+n]$$                           | Siqnal emalı |

Əlavə olaraq nümunə ilə fərqi riyazi göstərə bilərəm istəsən.


## Pareto prinsipi

Pareto prinsipi (və ya Pareto qanunu) — işlərin, nəticələrin və ya səbəblərin **təxminən 80%-i, səbəblərin və ya amillərin 20%-indən** qaynaqlandığını deyən müşahidədir.

### Qısaca:

* **80/20 qaydası** kimi də tanınır.
* Məsələn, bir şirkətin gəlirlərinin 80%-i, müştərilərin 20%-dən gəlir.
* Yaxud problemlərin 80%-i, səbəblərin 20%-indən yaranır.

### Tarixi:

Bu prinsip İtalyan iqtisadçısı Vilfredo Pareto tərəfindən 1896-cı ildə müşahidə olunub. O, İtaliyadakı torpaq paylanmasının təxminən 80%-nin, əhalinin 20%-inə aid olduğunu tapıb.

### İstifadə sahələri:

* İş idarəçiliyi
* Keyfiyyət nəzarəti
* Prioritetlərin müəyyənləşdirilməsi
* Problemlərin analizində

### Nümunə:

Bir layihədə ən kritik 20% tapşırıq, layihənin uğuruna 80% təsir edir.
