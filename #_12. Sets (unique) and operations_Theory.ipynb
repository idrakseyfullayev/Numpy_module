## Sets (unique) and operations_Theory

* **np.unique()** – Massivdəki unikal (təkrarsız) elementləri tapmaq üçün istifadə olunur.
  `unique_elements = np.unique(a)`

  Burada:

  * `a` – 1D və ya çoxölçülü massiv.
  * Qaytarılan `unique_elements` – `a` massivindəki təkrarsız elementlərin sıralanmış massividir.
  * İstəyə görə əlavə parametrlərlə elementlərin sayını, indekslərini də əldə etmək mümkündür (`return_counts=True`, `return_index=True` və s.).
  
---  

* **np.unique(a, return\_inverse=True)** – Massivdəki unikal elementləri tapmaqla yanaşı, orijinal massivdəki elementlərin unikal massivdəki indekslərini də qaytarır.
  `unique_elements, inverse_indices = np.unique(a, return_inverse=True)`

  Burada:

  * `a` – 1D və ya çoxölçülü massiv.
  * `unique_elements` – `a` massivindəki təkrarsız və sıralanmış elementlər.
  * `inverse_indices` – `a` massivinin hər bir elementinin `unique_elements` massivindəki indeksini göstərən massiv.
  * Bu, orijinal massivdəki elementləri unikal massivdəki indekslərə çevirir.

---

* **np.in1d()** – Bir massivdəki elementlərin başqa massivdə olub-olmadığını yoxlamaq üçün istifadə olunur.
  `result = np.in1d(a, b)`

  Burada:

  * `a` – yoxlanacaq massiv.
  * `b` – müqayisə üçün massiv.
  * Qaytarılan `result` – `a`-nın hər bir elementi üçün `b` massivində olub-olmamasını göstərən bool massividir (`True` və ya `False`).

---

* **np.intersect1d()** – İki massiv arasında kəsişən (ortaq) unikal elementləri tapmaq üçün istifadə olunur.
  `common_elements = np.intersect1d(a, b)`

  Burada:

  * `a` və `b` – müqayisə ediləcək massivlər.
  * Qaytarılan `common_elements` – hər iki massivdə olan unikal elementlərin sıralanmış massividir.

---

* **np.union1d()** – İki massivdəki bütün unikal elementləri birləşdirib sıralamaq üçün istifadə olunur.
  `all_unique = np.union1d(a, b)`

  Burada:

  * `a` və `b` – daxil olan massivlər.
  * Qaytarılan `all_unique` – `a` və `b` massivlərindəki bütün təkrarsız elementlərin sıralanmış massividir.

---

* **np.setdiff1d()** – Bir massivdə olub, digərində olmayan elementləri tapmaq üçün istifadə olunur.
  `difference = np.setdiff1d(a, b)`

  Burada:

  * `a` – əsas massiv.
  * `b` – müqayisə ediləcək massiv.
  * Qaytarılan `difference` – `a` massivində olub, `b` massivində olmayan unikal və sıralanmış elementlərdən ibarət massivdir.

---

* **np.setxor1d()** – İki massiv arasında yalnız birində olan unikal elementləri tapmaq üçün istifadə olunur.
  `symmetric_diff = np.setxor1d(a, b)`

  Burada:

  * `a` və `b` – müqayisə ediləcək massivlər.
  * Qaytarılan `symmetric_diff` – yalnız `a` və ya yalnız `b` massivində olan (hər ikisində eyni anda olmayan) unikal və sıralanmış elementlərdən ibarət massivdir.
